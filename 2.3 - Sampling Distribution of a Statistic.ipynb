#### ðŸ“˜ PART A: Conceptual Understanding

**Q1. Sampling Distribution Definition**
In your own words, what is a *sampling distribution*?
How is it different from the distribution of the population itself?

---

**Answer:** It is the distribution of some statistic (such as the mean) obtained from a number of samples drawn from a single population.

**Q2. Sample Size and Shape**
What happens to the shape and spread of the sampling distribution of the mean as the sample size increases?
Explain why this is important in statistical inference.

---

**Answer:** As the sample size increases, the distribution becomes more *normal* (develops more of a bell curve) even if the main data set doesn't skew towards the *normal*. This is important in statistical inference because it allows formulas like the t-distribution to be used in calculating sampling distributions for inference.

**Q3. Standard Error Calculation**
Given a population with standard deviation Ïƒ = 15 and a sample size of n = 25, calculate the standard error of the mean.

---

Formula:



```python
import math
std_err = 15 / math.sqrt(25)
print(f"The standard error is: {std_err}")
```

**Q4. Visual Interpretation**
Imagine you plot:

* Histogram A: 1,000 individual data values
* Histogram B: 1,000 means of samples with 5 values each
* Histogram C: 1,000 means of samples with 20 values each

Which histogram would be:

* Most bell-shaped?
* Narrowest?
* Most like the population?

Justify your choices.

---

**Answers:**
* Most bell-shaped: Histogram C, because the more means sampled, the more they tend to form the bell-shape according to the Central Limit Theorem.
* Narrowest: Histogram C, because the larger the sample, the narrower the distribution of the statistic.
* Most like the population: Histogram A, because individual data values tend to be more spread out and skewed toward higher values like the population.



**Q5. Practical Impact**
Why is the concept of the *sampling distribution* critical in estimating population parameters using sample statistics?

---

**Answers:** 

* Helps us estimate how much sample statistics (like means or proportions) are expected to vary

* Allows us to build confidence intervals, perform hypothesis tests, and make statistical inferences

## ðŸ’» Bonus Exercise: Simulating Sampling Distributions

### ðŸ”§ Setup:

Simulate a population of 100,000 loan applicants with **right-skewed incomes** (e.g., exponential distribution). Then:

### ### **Part A: Visualize the Population**

1. Simulate a population of 100,000 incomes with an exponential distribution:


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
population = np.random.exponential(scale=40000, size=100000)
population
```

2. Plot a histogram of the population income distribution.
   Does it look like a normal distribution?

---

```python
plt.hist(population, bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Distribution Histogram")
plt.show()
```

**Observation:** It doesn't look like a normal distribution because it is skewed to the right.

### **Part B: Simulate Sample Means**

1. Take **1,000 random samples** of **size 5** and compute the mean of each sample.


```python
# Initialize array to store the means
size_5_means = []

# Create loop to sample 1,000 times
for i in range(1000):
    sampled_values = np.random.choice(population, 5, replace=True)  # Get 5 random samples
    size_5_means.append(sampled_values.mean())  # save the mean of the 5 samples to our array
```

2. Repeat the process with **sample size 30**.

```python
# Initialize array to store the means
size_30_means = []

# Create loop to sample 1,000 times
for i in range(1000):
    sampled_values = np.random.choice(population, 30, replace=True)  # Get 30 random samples
    size_30_means.append(sampled_values.mean())  # save the mean of the 30 samples to our array
```

3. Plot histograms of both sets of sample means.

```python
plt.hist(size_5_means, bins=30, color='#CBC3E3', edgecolor='black')
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Sample Size of 5")
plt.show()
```

```python
plt.hist(size_30_means, bins=30, color='#CBC3E3', edgecolor='black')
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.title("Sample Size of 30")
plt.show()
```

4. Compare them in terms of:

   * Shape (normal or not?)
   * Spread (which is tighter?)

#### Answer:
The distribution of the graph with **Sample Size of 30** has a lower spread (it is tighter) and it's distribution is also more "normal" shaped than the one with **Sample Size of 5**. It's distribution tends more towards the **normal**.

### **Part C: Calculate Standard Error**

1. Compute the **standard deviation of the population**.
2. Compute the **standard deviation of the sample means** for size 5 and size 30.
3. Compare them to the theoretical **standard error formula**:


```python
pop_std_dev = population.std()
std_dev_5 = np.std(size_5_means)
std_dev_30 = np.std(size_30_means)

print(f"The standard deviation of the population is: {pop_std_dev}\n")
print(f"The standard deviation of the sample means for size 5 is: {std_dev_5}")
print(f"The standard error of the mean with sample size 5 is: {pop_std_dev / math.sqrt(5)}\n")
print(f"The standard deviation of the sample means for size 30 is: {std_dev_30}")
print(f"The standard error of the mean with sample size 30 is: {pop_std_dev / math.sqrt(30)}")
```

**Observation:** It is seems that the standard error is generally close in value to the standard deviation. The standard error shrinks with larger sample sizes and closely matches the theoretical prediction.
